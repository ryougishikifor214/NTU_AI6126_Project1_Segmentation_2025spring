```python
import os
import cv2
import numpy as np
from collections import defaultdict
from tqdm import tqdm

# æ©ç æ–‡ä»¶å¤¹è·¯å¾„
mask_dir = ORIGIN_ANNO_DIR_PATH

# å­˜å‚¨æ‰€æœ‰å”¯ä¸€çš„é¢œè‰²
unique_colors = set()

# éå†æ‰€æœ‰æ©ç æ–‡ä»¶
for filename in tqdm(os.listdir(mask_dir)):
    if filename.endswith(".png"):
        mask_path = os.path.join(mask_dir, filename)
        mask = cv2.imread(mask_path)  # è¯»å–ä¸º RGB å›¾åƒ (OpenCV é»˜è®¤æ˜¯ BGR)
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)  # è½¬æ¢ä¸º RGB
        unique_colors.update(map(tuple, mask.reshape(-1, 3)))  # è®°å½•å”¯ä¸€é¢œè‰²

# è¾“å‡ºæ‰€æœ‰å‘ç°çš„é¢œè‰²
unique_colors = sorted(unique_colors)  # æ’åºæ–¹ä¾¿æŸ¥çœ‹
print(f"å‘ç° {len(unique_colors)} ç§é¢œè‰²")
for color in unique_colors:
    print(color)
```
