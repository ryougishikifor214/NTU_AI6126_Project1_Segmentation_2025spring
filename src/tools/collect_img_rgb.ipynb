```python
import os
import cv2
import numpy as np
from collections import defaultdict
from tqdm import tqdm

# 掩码文件夹路径
mask_dir = ORIGIN_ANNO_DIR_PATH

# 存储所有唯一的颜色
unique_colors = set()

# 遍历所有掩码文件
for filename in tqdm(os.listdir(mask_dir)):
    if filename.endswith(".png"):
        mask_path = os.path.join(mask_dir, filename)
        mask = cv2.imread(mask_path)  # 读取为 RGB 图像 (OpenCV 默认是 BGR)
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)  # 转换为 RGB
        unique_colors.update(map(tuple, mask.reshape(-1, 3)))  # 记录唯一颜色

# 输出所有发现的颜色
unique_colors = sorted(unique_colors)  # 排序方便查看
print(f"发现 {len(unique_colors)} 种颜色")
for color in unique_colors:
    print(color)
```
